```python
from math import factorial
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def calc_theta(n_h, n_t):
    return (n_h + 1) / (n_h + n_t + 2)

def calc_prob(n_h, n_t):
    tetta = (n_h + 1) / (n_h + n_t + 2)
    return tetta ** n_h * (1 - tetta) ** n_t / (factorial(n_h) * factorial(n_t) / factorial(n_h + n_t + 1))

def calc_prob_1(n_h, n_t):
    if n_h + n_t  == 0:
        return 0
    tetta = (n_h ) / (n_h + n_t )
    
    return tetta ** n_h * (1 - tetta) ** n_t / (factorial(n_h) * factorial(n_t) / factorial(n_h + n_t + 1))

clc = np.vectorize(calc_prob)
clh = np.vectorize(calc_theta)
clc1 = np.vectorize(calc_prob_1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n_h_ = np.linspace(0, 50, 51)
n_t_ = np.linspace(0, 50, 51)
m = np.meshgrid(n_h_, n_t_)
z = clh(m[0], m[1])

ax.plot_surface(m[0], m[1], z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
z1 = clc1(m[0], m[1])

ax.plot_surface(m[0], m[1], z1)
plt.show()
```
