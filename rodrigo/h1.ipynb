```python
import random
import numpy as np

import matplotlib.pyplot as plt
```

```python
def prob_heads(Nh, Nt):
    return (Nh + 1)/(Nh + Nt + 2)
```

## Equal probability


```python
p_heads = 0.5
Nh = 0
Nt = 0
tot_probs = []

for i in range(20000):
    prob = random.uniform(0,1)
    if prob <= p_heads:
        Nh += 1
    else:
        Nt += 1
    tot_probs.append(prob_heads(Nh, Nt))

plt.figure(figsize=(12,4))
plt.plot(tot_probs)
plt.show()
```

## Higher probability for heads

```python
p_heads = 0.7
Nh = 0
Nt = 0
tot_probs = []

for i in range(20000):
    prob = random.uniform(0,1)
    if prob <= p_heads:
        Nh += 1
    else:
        Nt += 1
    tot_probs.append(prob_heads(Nh, Nt))

plt.figure(figsize=(12,4))
plt.plot(tot_probs)
plt.show()
```
